```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def iterate(A,B):

    d = math.dist(A,B)

    point_1 = (A[0],A[1])

    point_2 = ( 
        A[0]+(B[0]-A[0])/3, 
        A[1]+(B[1]-A[1])/3
    )
    
    angle = np.arctan2((B[1]-A[1]),(B[0]-A[0]))
    
    point_3 = (
        point_2[0] + np.cos(((np.pi/3)+angle))*d/3,
        point_2[1] + np.sin(((np.pi/3)+angle))*d/3
    )

    point_4 = ( 
        A[0]+2*(B[0]-A[0])/3, 
        A[1]+2*(B[1]-A[1])/3
    )

    point_5 = (B[0],B[1])

    return [point_1, point_2, point_3, point_4, point_5]
```

```python
STEPS = 4
initial_points = [(0,0),(1,0)]

texts = ['(a)', '(b)', '(c)', '(d)', '(e)']

xs = [coords[0] for coords in initial_points]
ys = [coords[1] for coords in initial_points]

f,ax =plt.subplots(STEPS+1,1,figsize=(5,2.5*(STEPS+1)))    
ax[0].plot(xs,ys, c='k')
ax[0].set_xlim(-0.1,1.1)
ax[0].set_ylim(-0.1,0.55)

for step in range(STEPS):
    new_points = [initial_points[0]]
    for pair in range(len(initial_points)-1):
        new_points.extend(iterate(initial_points[pair], initial_points[pair+1])[1:])
    initial_points = new_points

    xs = [coords[0] for coords in initial_points]
    ys = [coords[1] for coords in initial_points]

    ax[step+1].plot(xs,ys, c='k')
    ax[step+1].set_ylim(-0.1,0.55)

for i in range(5):
    ax[i].text(0.9, 0.9, texts[i], transform=ax[i].transAxes)

f.subplots_adjust(wspace=0, hspace=0)
plt.savefig('D:/Workshop/PhD/Thesis/Chapter 2/KochCurve.png', bbox_inches = 'tight')
plt.show()
```
